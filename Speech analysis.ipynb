# Problem

In this particular project, we are going to work on the inaugural corpora from the nltk in Python. We will be looking at the following speeches of the Presidents of the United States of America:

```python
import numpy as np
import pandas as pd
import nltk
from nltk.stem.snowball import SnowballStemmer
import json
import re
import string
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud,STOPWORDS
import pyphen
nltk.download('stopwords')
nltk.download('punkt')
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
name = inaugural.fileids()
```

```python
word = inaugural.raw(fileids = name[38])
word
```

```python
len(inaugural.raw(fileids = name[38]))
```

```python
nltk.FreqDist(inaugural.words(fileids = name[38])).most_common(10)
```

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)
stemmer = SnowballStemmer("english")
all_words = (w.lower() for w in inaugural.words(fileids = name[38])) 
all_words
all_words_clean = [word for word in all_words if word not in stopwords]
all_words_clean
```

```python
len(all_words_clean)
```

```python
df = pd.DataFrame(all_words_clean)
df.value_counts()
```

```python
text = inaugural.raw(fileids = name[38])
wordcloud = WordCloud(max_font_size=60).generate(text)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,12))
plt.imshow(wordcloud, interpolation="bilinear")

plt.axis("off")
plt.show()
```
